```python
import cv2
import time
import numpy as np
#import sys

# Method 8 HSV to Gray
# https://www.geeksforgeeks.org/python-opencv-cv2-cvtcolor-method/

vidCap = cv2.VideoCapture('B3-Parking.avi')

# read the frame 1
ret, frame1 = vidCap.read()
    
frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2HSV)
cv2.imshow('Window1', frame1)

while vidCap.isOpened():

    ret, frame2 = vidCap.read()
    
    if not ret:
        break
        
    frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2HSV)
    cv2.imshow('Window2', frame2)

    # extract the foreground frame
    fgMask = cv2.absdiff(frame1, frame2)
    # cv2.imshow('Foreground Mask', fgMask)
    
    # Apply the threshold
    
    _, thres = cv2.threshold(fgMask, 25, 255, cv2.THRESH_BINARY) # Threshold 25 because of the colors of cars
    cv2.imshow('Threshold', thres)
    

    # assign frame2 to frame1 to continue the iteration until all frames are read
    frame1 = frame2

    if cv2.waitKey(10) & 0xFF == 27:
        break


#  releasse video capture
cv2.destroyAllWindows()
vidCap.release()
```

```python
import cv2
import time
#import sys

# Method 5 Median Blurring

vidCap = cv2.VideoCapture('B3-Parking.avi')

# read the frame 1
ret, frame1 = vidCap.read()
    
frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
cv2.imshow('Window1', frame1)

while vidCap.isOpened():

    ret, frame2 = vidCap.read()
    
    if not ret:
        break
        
    frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Window2', frame2)

    # extract the foreground frame
    fgMask = cv2.absdiff(frame1, frame2)
    
    blur = cv2.medianBlur(fgMask, 5) # Median Blurring

    # Apply the threshold
    _, thres = cv2.threshold(blur, 25, 255, cv2.THRESH_BINARY) # Threshold 25 because of the colors of cars
    cv2.imshow('Median Blur', thres)
    
    # assign frame2 to frame1 to continue the iteration until all frames are read
    frame1 = frame2

    if cv2.waitKey(10) & 0xFF == 27:
        break


#  releasse video capture
cv2.destroyAllWindows()
vidCap.release()
```

```python
import cv2
import time
#import sys

# Method 3  Blur

vidCap = cv2.VideoCapture('B3-Parking.avi')

# read the frame 1
ret, frame1 = vidCap.read()
    
frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
cv2.imshow('Window1', frame1)

while vidCap.isOpened():

    ret, frame2 = vidCap.read()
    
    if not ret:
        break
        
    frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Window2', frame2)

    # extract the foreground frame
    fgMask = cv2.absdiff(frame1, frame2)
    
    blur = cv2.blur(fgMask,(7, 7))

    # Apply the threshold
    _, thres = cv2.threshold(blur, 25, 255, cv2.THRESH_BINARY) # Threshold 25 because of the colors of cars
    cv2.imshow('Blur', thres)
    
    # assign frame2 to frame1 to continue the iteration until all frames are read
    frame1 = frame2

    if cv2.waitKey(30) & 0xFF == 27:
        break


#  releasse video capture
cv2.destroyAllWindows()
vidCap.release()
```

```python
# Method 1

vidCap = cv2.VideoCapture('B3-Parking.avi')

# read the frame 1
ret, frame1 = vidCap.read()
    
frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
cv2.imshow('Window1', frame1)

while vidCap.isOpened():

    ret, frame2 = vidCap.read()
    
    if not ret:
        break
        
    frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Window2', frame2)

    # extract the foreground frame
    fgMask = cv2.absdiff(frame1, frame2)

    # Apply the threshold
    _, thres = cv2.threshold(fgMask, 30, 255, cv2.THRESH_BINARY)
    cv2.imshow('ForegroundMask', thres)
    
    # assign frame2 to frame1 to continue the iteration until all frames are read
    frame1 = frame2

    if cv2.waitKey(30) & 0xFF == 27:
        break


#  releasse video capture
cv2.destroyAllWindows()
vidCap.release()
```

```python
# Mrthod 2

vidCap = cv2.VideoCapture('B3-Parking.avi')

BS_KNN = cv2.createBackgroundSubtractorKNN()
BS_MOG2 = cv2.createBackgroundSubtractorMOG2()

while vidCap.isOpened():
    ret, frame = vidCap.read() # read the next frame
    
    # extract the KNN-method of foreground Mask
    knn_FGMask = BS_KNN.apply(frame)
    cv2.imshow('KNN', knn_FGMask)

    # extract the MOG2-method of foreground Mask
    mog2_FGMask = BS_MOG2.apply(frame)
    cv2.imshow('MOG2', mog2_FGMask)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break

#  releasse video capture
cv2.destroyAllWindows()
vidCap.release()
```
